```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
armadillo_noAccel = 229.46
bunny_noAccel = 90.9341
lucy_noAccel = 199.992
all_three_noAccel = 868.339

bunny_accel = 0.591368
armadillo_accel = 0.836735  
lucy_accel = 0.872304
all_three_accel = 1.53416

triangles_armadillo = 3112
triangles_bunny = 1392
triangles_lucy = 2804
triangles_all = triangles_bunny + triangles_armadillo + triangles_lucy


num_triangles = [triangles_bunny, triangles_armadillo, triangles_lucy, triangles_all]

render_times_noAccel = [bunny_noAccel, armadillo_noAccel, lucy_noAccel, all_three_noAccel]
render_times_accel = [bunny_accel, armadillo_accel, lucy_accel, all_three_accel]


# Create a scatter plot for both models
plt.scatter(num_triangles, render_times_noAccel, color='blue', label='No acceleration', marker='o')
plt.scatter(num_triangles, render_times_accel, color='red', label='Acceleration (kd-tree)', marker='x')

# Optional: Fit and plot regression lines for each model
coefficients_model1 = np.polyfit(num_triangles, render_times_noAccel, 2)
polynomial_model1 = np.poly1d(coefficients_model1)
plt.plot(num_triangles, polynomial_model1(num_triangles), color='blue', linestyle='--')

coefficients_model2 = np.polyfit(num_triangles, render_times_accel, 2)
polynomial_model2 = np.poly1d(coefficients_model2)
plt.plot(num_triangles, polynomial_model2(num_triangles), color='red', linestyle='--')
# Set logarithmic scale on both axes
# plt.xscale('log')
plt.yscale('log')

# Add titles and labels
plt.title("Render Time Comparison: No acceleration vs Acceleration (kd-tree)")
plt.xlabel("Number of Triangles")
plt.ylabel("Render Time (seconds)")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
